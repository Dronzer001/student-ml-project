```python
import pandas as pd
from sklearn.linear_model import LogisticRegression
import joblib

# Load data
df = pd.read_csv("data/student.csv")

X = df[['hours']]
y = df['pass']

# Train model
model = LogisticRegression()
model.fit(X, y)

# Save model
joblib.dump(model, "model.joblib")

print("Model trained and saved!")
```
